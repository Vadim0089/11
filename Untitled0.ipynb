```python
import torch
from diffusers import __version__ as diffusers_version
from transformers import __version__ as transformers_version

print(f"PyTorch version: {torch.__version__}")
print(f"Diffusers version: {diffusers_version}")
print(f"Transformers version: {transformers_version}")
```

```python
# Установка библиотек
!pip install diffusers transformers torch

# Импорт библиотек
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image

# Проверка версий
print(f"PyTorch version: {torch.__version__}")

# Задаем модель для генерации изображений
model_id = "runwayml/stable-diffusion-v1-5"

# Создаем пайплайн для генерации изображений
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)

# Переводим модель на GPU
device = "cuda"
pipe = pipe.to(device)

# Включаем attention slicing для экономии памяти
pipe.enable_attention_slicing()

# Текстовое описание для генерации изображения
prompt = "The car"

# Генерируем изображение
image = pipe(prompt, height=256, width=256).images[0]

# Отображаем изображение
display(image)

# Сохраняем изображение
image.save("generated_image.png")
print("Изображение успешно сохранено!")
```
